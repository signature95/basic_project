```python
# data 폴더에 있는 모든 csv 파일을 읽어오기 위해 glob을 사용합니다.
from glob import glob

# csv 목록 불러오기
file_names = glob('소상공인시장진흥공단_상가(상권)정보/*.csv')
total = pd.DataFrame()
# print(file_names)
# pd.concat : csv파일 데이터를 이어서 합치기
for file_name in file_names:
    df = pd.read_csv(file_name, sep=',', encoding='utf-8')
    total = pd.concat([total, df])

total

# 모든 csv 병합하기
```

```python
# reset index
total.reset_index(inplace=True, drop=True)

total
```

### 1) 전체 커피전문점 내 주요 커피브랜드 입점 비율 

```python
num_coffee = len(df_coffee)
num_starbucks = len(df_starbucks)
num_2some = len(df_2some)
num_ediya = len(df_ediya)
num_mega = len(df_mega)
num_coffeebean = len(df_coffeebean)
num_bbaek = len(df_bbaeck)
num_hollys = len(df_hollys)


num_seoul_coffee = len(df_seoul_coffee)
num_seoul_starbucks = len(df_seoul_starbucks)
num_seoul_2some = len(df_seoul_2some)
num_seoul_ediya = len(df_seoul_ediya)
num_seoul_mega = len(df_seoul_mega)
num_seoul_coffeebean = len(df_seoul_coffeebean)
num_seoul_bbaek = len(df_seoul_bbaeck)
num_seoul_hollys = len(df_seoul_hollys)
```

```python
# 시각화를 위한 라이브러리를 불러옵니다.
import seaborn as sns
import matplotlib.pyplot as plt
import platform

from matplotlib import font_manager, rc
%matplotlib inline
```

```python
#### Windows10 사용자는 해당 코드가 필요없습니다!

# macos에서 사용가능한 한글 글꼴 확인 코드
[f.name for f in font_manager.fontManager.ttflist if 'Neo' in f.name]
```

```python
## 운영체제별 글꼴 세팅

path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin':
    font_name = 'Apple SD Gothic Neo'
    rc('font', family='Apple SD Gothic Neo')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    font_name = font_manager.FontProperties(fname="/usr/share/fonts/nanumfont/NanumGothic.ttf")
    rc('font', family="NanumGothic")
```

```python
# 주요 5대 커피브랜드 서울 입점 비율을 시각화합니다.
starbucks_rate = (num_seoul_starbucks / num_seoul_coffee)*100
ediya_rate = (num_seoul_ediya / num_seoul_coffee)*100
hollys_rate = (num_seoul_hollys / num_seoul_coffee)*100
twosome_rate = (num_seoul_2some / num_seoul_coffee)*100
mega_rate = (num_seoul_mega / num_seoul_coffee)*100

X = ['스타벅스','투썸플레이스', '이디야','할리스','메가커피']
Y = [starbucks_rate, twosome_rate, ediya_rate, hollys_rate, mega_rate]

plt.figure(figsize=(8,8))
plt.title('주요 5대 커피브랜드 서울 입점 비율', fontdict={'fontsize' : 20})
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
sns.barplot(x= X, y= Y)
plt.show()
```
